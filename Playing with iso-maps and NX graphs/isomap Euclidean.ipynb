```python
# calculating the threshold.
```

```python
l=pd.DataFrame(eu, index=None)
```

```python
l.head()
```

```python
l.sort_values(by=0, axis=1, ascending=True, inplace=True)
```

```python
l.head()
```

```python
lst=[]

for i in range(698):
    l.sort_values(by=i, axis=1, ascending=True, inplace=True)
    lst.append(l.iloc[i,100])

np.max(lst)
```

```python
# minimum threshhold for having 100 neighbours for each node is 22.388619203990206 distance.
# we'll set this as a bench mark for similarity
```

```python
h={}
for i in range(1,698):
    h[i]=[]       # creating an empty dictionary to further store the connected nodes for each node with the set threshold level.
```

```python
for i in range(1,698):
    for j in range(1, 698):
        if l[i][j]<=22.388619203990206:
            h[i].append(j)     #creating the dictionary.
```

```python
gnew = nx.Graph(h)      #plotting the graph with the dictionary.
```

```python
nx.draw(gnew)
```

```python
k1=nx.adjacency_matrix(gnew)
```

```python
#implementing the Isomap with the adjacency matrix got above to reduce the dimensions.
```

```python
from sklearn import manifold
```

```python
iso2 = manifold.Isomap(n_neighbors=100, n_components=2)
iso2.fit(k1)
```

```python
Z=k1.toarray()
manifold_2Dan = iso2.transform(Z)
manifold_2Dn = pd.DataFrame(manifold_2Dan, columns=['Component 1', 'Component 2'])

# Left with 2 dimensions
manifold_2Dn.head()
```

```python
num_images, num_pixels = df.shape
```

```python
import math
pixels_per_dimensionn = int(math.sqrt(num_pixels))

# Rotate the pictures
for idx in df.index:
    df.loc[idx] = df.loc[idx].values.reshape(pixels_per_dimensionn, pixels_per_dimensionn).T.reshape(-1)
    
# Show first 5 rows
print(df.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111)
ax.set_title('2D Components from Isomap of Facial Images')
ax.set_xlabel('Component: 1')
ax.set_ylabel('Component: 2')

# Show 40 of the images ont the plot
x_size = (max(manifold_2Dn['Component 1']) - min(manifold_2Dn['Component 1'])) * 0.08
y_size = (max(manifold_2Dn['Component 2']) - min(manifold_2Dn['Component 2'])) * 0.08
for i in range(40):
    img_num = np.random.randint(0, num_images)
    x0 = manifold_2Dn.loc[img_num, 'Component 1'] - (x_size / 2.)
    y0 = manifold_2Dn.loc[img_num, 'Component 2'] - (y_size / 2.)
    x1 = manifold_2Dn.loc[img_num, 'Component 1'] + (x_size / 2.)
    y1 = manifold_2Dn.loc[img_num, 'Component 2'] + (y_size / 2.)
    img = df.iloc[img_num,:].values.reshape(pixels_per_dimensionn, pixels_per_dimensionn)
    ax.imshow(img, aspect='auto', cmap=plt.cm.gray, 
              interpolation='nearest', zorder=100000, extent=(x0, x1, y0, y1))

# Show 2D components plot
ax.scatter(manifold_2Dn['Component 1'], manifold_2Dn['Component 2'], marker='.',alpha=0.7)

ax.set_ylabel('Up-Down Pose')
ax.set_xlabel('Right-Left Pose')

plt.show()
```
