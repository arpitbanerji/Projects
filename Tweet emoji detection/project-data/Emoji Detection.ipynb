```python
lst=[]
for i in text1[2]:    
    k=i.split()
    if k[0]=='rt':
        k.remove('rt')
    lst.append(k)                #removing the rt word
```

```python
text1[2]=lst
```

```python
lst=[]
for i in text1[2]:
    lst.append(" ".join(i))
```

```python
text1[2]=lst
```

```python
stop = stopwords.words("english") #removing stopwords
text1[2]=text1[2].str.split()
def removestopwords(y):   
    stopwordremoved =[w for w in y if not w in stop]
    return(" ".join(stopwordremoved))
column_size = text1.shape[0]
  
lst = []
for i in range( 0, column_size):
    lst.append(removestopwords(text1[2][i]))
text1[2]=lst
```

```python
d={}    #creating a dictionary of vocablury
for i in text1[2]:
    k=i.split()
    for m in k:
        if m in d.keys():
            d[m]=d[m]+1
        if m not in d.keys():
            d[m]=1
```

```python
d   #vocab
```

```python
len(d)
```

```python
sorted_d={k: v for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)}
```

```python
sorted_d     #we can drop the most repeating words from the vocab like 'https''t' 'co','i', 'the'
```

```python
lst=[]
for i in text1[2]:    
    k=i.split()
    for j in k:  
        for i in sorted_d:
            if j==i :
                k.remove(j)
            if d[i]<1000:
                break;      
    lst.append(k)
```

```python
text1[2]=lst
```

```python
lst=[]
for i in text1[2]:
    lst.append(" ".join(i))
```

```python
text1[2]=lst
```
